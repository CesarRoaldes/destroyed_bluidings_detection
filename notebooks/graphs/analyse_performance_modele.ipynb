# Evolution des performance (loss / accuracy) lors de l'apprentissage

On charge les informations de la phase d'apprentissage précédement enregistré en format pickle.

```python
with open('../../data/model/all_model_no_euroSAT_info.pkl', 'rb') as handle:
    all_model_no_eurosat_info = pickle.load(handle)
```

```python
with open('../../data/model/all_model_euroSAT_info.pkl', 'rb') as handle:
    all_model_eurosat_info = pickle.load(handle)
```

```python
nb_epoch = len(all_model_eurosat_info['train']['loss'])
```

## Analyse graphique de l'accuracy et de la loss sur les deux modèles

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot([float(i)  for i in all_model_eurosat_info['test']['accuracy']], 'b', label='pretrained sur euroSAT')
ax.plot([float(i)  for i in all_model_no_eurosat_info['test']['accuracy']], 'orange', label='non pretrained')
ax.legend(loc='center right', fontsize=15)
ax.set_xticks(np.arange(0,51,5))
ax.set_title("Evolution de l'accuracy par epoch", fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot([float(i)  for i in all_model_eurosat_info['test']['loss']], 'b', label='pretrained sur euroSAT')
ax.plot([float(i)  for i in all_model_no_eurosat_info['test']['loss']], 'orange', label='non pretrained')
ax.legend(loc='center right', fontsize=15)
ax.set_xticks(np.arange(0,51,5))
ax.set_title('Evolution de la loss par epoch', fontsize=15)
plt.show()
```
